```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, HTML
sns.set_style("darkgrid")
```

```python
def create_scrollable_table(df, table_id, title):
    html = f'<h3>{title}</h3>'
    html += f'<div id="{table_id}" style="height:200px; overflow:auto;">'
    html += df.to_html()
    html += '</div>'
    return html
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
numerical_features = df.select_dtypes(include=[np.number])
summary_stats = numerical_features.describe().T
html_numerical = create_scrollable_table(summary_stats, 'numerical_features', 'Summary statistics for numerical features')
display(HTML(html_numerical))
```

```python
# Null values in the dataset
null_values = df.isnull().sum()
html_null_values = create_scrollable_table(null_values.to_frame(), "null_values", "Null values in the dataset")

#Percentage of missing values for each feature
missing_percentage = (df.isnull().sum() / len(df)) * 100
html_missing_values = create_scrollable_table(missing_percentage.to_frame(), "missing_percentage", "Percentage of missing value for each feature")

display(HTML(html_null_values + html_missing_values))
```

# Exploring rows with missing values

```python
row_with_missing_values = df[df.isnull().any(axis=1)]
html_row_missing_values = create_scrollable_table(row_with_missing_values.head(), "rows_with_missing_values", "Rows with missing values")
display(HTML(html_row_missing_values))
```
